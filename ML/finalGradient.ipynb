```python
import numpy as np
import pandas as pd
from matplotlib import pyplot
```

```python
def objective(x):
    return (x+3)**2
```

```python
def derivative(x):
    return 2*(x+3)
```

```python
def gradient(alpha,max_itr,start):
    x_list=list()
    x=start
    x_list.append(x)
    for i in range (max_itr):
        gradi=derivative(x)
        x=x-(alpha*gradi)
        x_list.append(x)
    return x_list    
```

```python
alpha=0.1
max_itr=30
start=2
x_l=gradient(alpha,max_itr,start)
x_array=np.array(x_l)
# pyplot.plot(x_array,objective(x_array),'.-',color='red')
pyplot.plot(x_array,objective(x_array),'.-',color='purple')
```
